```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


# Load the CSV file into a DataFrame
csv_file = "/content/Book1.csv"
df = pd.read_csv(csv_file)

```

```python
df.head()
```

```python
# Filter the data for any Country
country_data = df[df['Countries'] == 'Albania']

# Get the age distribution percentages
age_groups = ['Ages 0-14(%)', 'Ages 15-64(%)', 'Ages 65+(%)']
age_distribution = country_data[age_groups].values[0]

# Define custom colors for the bars
custom_colors = ['#9B59B6', '#2ECC71', '#D35400']

# Create a bar graph for the age distribution with custom colors
plt.figure(figsize=(8, 6))
bars = plt.bar(age_groups, age_distribution, color=custom_colors)
plt.xlabel('Age Group')
plt.ylabel('% Population')
plt.title(' Country-Wise Age Distribution')
plt.ylim(0, 100)  # Set the y-axis limits to 0-100

# Add color legend
for bar, age_group in zip(bars, age_groups):
    bar.set_label(age_group)

plt.legend(title='Age Groups', loc='upper right')
plt.show()
```
